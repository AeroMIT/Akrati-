```python
#how to filter a video to a specific range of color
#HSV has better color detection ,color remain stable in different lighting(hue remain stable)
#MASK : is used to specify a region of interest in an image, allowing operation to be performed only on selected area
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

#if not cap.isOpened():  # Check if the camera is accessible
    #print("Error: Could not open camera")
    #exit()
    
while True:
    ret, frame = cap.read()
    #if not ret:
        #print("Failed to grab frame")
        #break
    
    #hsv is hue sat value (HSV) color frame 
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)#cv2.COLOR_BGR2HSV convert BGR to HSV

    lower_pink = np.array([150,150,50]) #define lower boundry of the color selected
    upper_pink = np.array([180,100,255])#define upper boundry of the color selected

    mask = cv2.inRange(hsv, lower_pink, upper_pink) #mask stores the resulting binary image
    res = cv2.bitwise_and(frame, frame, mask = mask) #res stores the resulting image; bitwise_and() allows us to extract only the desired colored objects from the image.

    cv2.imshow('frame', frame)
    cv2.imshow('mask', mask) 
    cv2.imshow('res', res)

    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#SMOOTHING TECHNIQUES
#KERNEL : is a small matrix used to apply effects such as blurring,image filtering,edge detection,noise reduction.
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
     
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower_pink = np.array([150, 50, 180]) 
    upper_pink = np.array([180, 100, 255])

    mask = cv2.inRange(hsv, lower_pink, upper_pink) 
    res = cv2.bitwise_and(frame, frame, mask = mask) 

    #SMOOTHING
    kernel = np.ones((15,15), np.float32)/225
    smoothed = cv2.filter2D(res, -1, kernel)
    
    cv2.imshow('frame', frame)
    #cv2.imshow('mask', mask) 
    cv2.imshow('res', res)
    cv2.imshow('smoothed', smoothed)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#BLUR TECHNIQUES
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
     
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower_pink = np.array([150,150,50]) 
    upper_pink = np.array([180,100,255])

    mask = cv2.inRange(hsv, lower_pink, upper_pink)  
    res = cv2.bitwise_and(frame, frame, mask = mask) 

    kernel = np.ones((15,15), np.float32)/225
    smoothed = cv2.filter2D(res, -1, kernel)

    blur = cv2.GaussianBlur(res, (15,15), 0) #best for smoothing images without losing edges.(image, kernel_size, sigmaX)
    median = cv2.medianBlur(res, 15) #removing random black and white dots(image, kernel_size)
    bilateral = cv2.bilateralFilter(res, 15,75,75)#remove noise while keeping the edge sharpe(image , daimeter of the pixel neighbourhood, color deviatio(standard and spatial))
    
    cv2.imshow('frame', frame)
    #cv2.imshow('mask', mask) 
    #cv2.imshow('res', res)
    #cv2.imshow('smoothed', smoothed)
    cv2.imshow('blur', blur)
    cv2.imshow('median', median)
    cv2.imshow('bilateral', bilateral)
    
    

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#Morphological Transformations : used to refine object structure. uses structuring element.used for(noise removal,extracting important shape)
#EROSION AND DILATION : E- remove pixel from object boundary. D- adds pixels to object boundaries.
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
     
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower_pink = np.array([150, 50, 180]) 
    upper_pink = np.array([180,100,255])

    mask = cv2.inRange(hsv, lower_pink, upper_pink)  
    res = cv2.bitwise_and(frame, frame, mask = mask) 

    kernel = np.ones((5,5), np.uint8)
    erosion = cv2.erode(mask, kernel, iterations = 1)
    dilation = cv2.dilate(mask, kernel, iterations = 1)
   
    cv2.imshow('frame', frame)
    cv2.imshow('res', res)
    cv2.imshow('erosion', erosion)
    cv2.imshow('dilation', dilation)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#opening : erosion followed by dilation. removes noise while keeping the main object shape
#closing: dilation followed by erosion. closes small holes inside objects
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower_pink = np.array([150, 50, 180]) 
    upper_pink = np.array([180,100,255])

    mask = cv2.inRange(hsv, lower_pink, upper_pink)  
    res = cv2.bitwise_and(frame, frame, mask = mask) 

    kernel = np.ones((5,5), np.uint8)
    opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    closeing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
   
    cv2.imshow('frame', frame)
    cv2.imshow('res', res)
    cv2.imshow('opening', opening)
    cv2.imshow('closeing', closeing)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#EDGE DETECTION 
#LAPLACIAN : used in image edge detection to detect edge of the image. highlight area of rapid intensity change.
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    laplacian = cv2.Laplacian(frame, cv2.CV_64F) #cv2.CV_64F specify output image length.

    cv2.imshow('original',frame)
    cv2.imshow('laplacian',laplacian)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#sobelx and sobely : calculates the gradient of the image intensity in a specific direction(horizontal or vertical)
# first is in X directio and second is in y-direction
#CANNY : detected edge are white and everthing else is black
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    sobelx = cv2.Sobel(frame, cv2.CV_64F, 1,0,ksize = 5)
    sobely = cv2.Sobel(frame, cv2.CV_64F, 0,1,ksize = 5)

    edges = cv2.Canny(frame, 100, 100) # lower and upper threshold value

    cv2.imshow('original',frame)
    cv2.imshow('sobelx',sobelx)
    cv2.imshow('sobely',sobely)
    cv2.imshow('edges',edges)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#TEMPLET MATCHING
#cv2.TM_CCOEFF_NORMED : It computes the correlation between the template and the image at different positions.
import cv2
import numpy as np

img_bgr = cv2.imread('opencv-template-matching-python-tutorial.jpg')
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

template = cv2.imread('opencv-template-for-matching.jpg', 0) # by passing 0 we are converting it to grayscale
w, h = template.shape[::-1] #returns (height, width)

res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED) #cv2.matchTemplate scans the first image for area that matchs the template.
threshold = 0.90
loc = np.where(res >= threshold)

for pt in zip(*loc[::-1]):
    cv2.rectangle(img_bgr, pt, (pt[0]+w, pt[1]+h), (0,255,255),2)

cv2.imshow('detected', img_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#GrabCut Foreground Extraction:Foreground extraction is essential in computer vision and image processing for isolating an object from its background.OpenCV's grabCut algorithm provides an efficient way to achieve this using graph-based segmentation. 
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('foreground-tutorial.jpg')
mask = np.zeros(img.shape[:2], np.uint8)#img.shape[:2] extracts the height and width of the image.The mask is later used to differentiate foreground and background regions.


bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)

rect = (50, 50, 300, 300)#(x, y, width, height)

cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask[:,:,np.newaxis]
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
#CORNER DETECTION
import cv2
import numpy as np

img = cv2.imread('opencv-corner-detection-sample.jpg')
if img is None:
    print("Error: Image not found or path incorrect")
    exit()
    
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#converting to grayscale because color detection works best on single-channel image.
gray = np.float32(gray)

corners = cv2.goodFeaturesToTrack(gray,100,0.01,10)
corners = np.int0(corners)#Converts detected corner coordinates from float to integer.

#corners = np.int0(corners)

for corner in corners:
    x,y = corner.ravel() #.ravel(): Flattens the array to extract x, y coordinates.
    cv2.circle(img,(x,y),3,255,-1)

cv2.imshow('Corner', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Feature Matching (Homography) Brute Force 
#Hamming distance is a measure of similarity between two binary strings.
#ORB (Oriented FAST and Rotated BRIEF) is a feature detector and descriptor used in computer vision, particularly for object recognition.
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('opencv-feature-matching-template.jpg',1)
img2 = cv2.imread('opencv-feature-matching-image.jpg',1)
if img1 is None or img2 is None:
    print("Error: One or both images not found.")
    exit()

orb = cv2.ORB_create() #Initializes the ORB (Oriented FAST and Rotated BRIEF) feature detector.

kp1, des1 = orb.detectAndCompute(img1,None)#Finds keypoints (distinctive points in an image, such as corners, edges, or blobs).
kp2, des2 = orb.detectAndCompute(img2,None)#Computes descriptors (numeric vectors describing the local patterns around keypoints).

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)

matches = bf.match(des1, des2)#Finds the best match for each descriptor in img1 from the descriptors in img2.
matches = sorted(matches, key = lambda x:x.distance)
img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:15],None,flags=2)#flags=2: Draws only the matched keypoints, ignoring unmatched ones.
plt.imshow(img3)
plt.show()
```

```python
#COUNTOR
import cv2
import numpy as np

img = cv2.imread('fruit.jpg')
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_purple = np.array([125,50,50])
upper_purple = np.array([160,255,255])

mask1 = cv2.inRange(img1, lower_purple, upper_purple)

kernel = np.ones((5,5),np.uint8)
opening1 = cv2.morphologyEx(mask1, cv2.MORPH_OPEN, kernel)

res1 = cv2.bitwise_and(img, img, mask=opening1)

gray = cv2.cvtColor(res1, cv2.COLOR_BGR2GRAY)

retval, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(img, contours, -1, (0, 255, 0), 3) 

cv2.imshow("Contours", img)
cv2.imshow('res1', res1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
